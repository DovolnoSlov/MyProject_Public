### Задание
#### Необходимо развернуть массивы данных с листа Исх.данные в строки, так как показано на листе Результат. 
#### Порядок строк должен соответствовать порядку элементов в массивах.

```python
import pandas as pd
import numpy as np
import re
from ast import literal_eval
```

```python
# Загрузка данных в датасет
dft = pd.read_excel("Задание_2.xlsx", sheet_name="Исх.данные")
```

```python
# замена булевого значения на число, для корректной работы literal_eval
dft['owner'] = dft['owner'].apply(lambda x: x.replace("true", "1"))
```

```python
# конвертация строк в списки
columns_dft = dft.columns[2:]
for col in columns_dft:
    dft[col] = dft[col].apply(literal_eval)
```

```python
# "раскрытие" списков в строки в каждом столбце с сопоставлением значений из других столбцов
# (сохраняя номера индексов при раскрытии)
# в случае указания нескольких столбцов - будет перекрёстное раскрытие без дублирования
dft = dft.explode(list(columns_dft))
# возвращение формата значений столбца "owner"
dft['owner'].replace(1, True, inplace=True)
dft
```

```python
# сохранение результата в excel файл, без учёта индексов, для получения требуемого формата вывода данных
dft.to_excel("result2.xlsx", sheet_name="Результат", index=False)
```
